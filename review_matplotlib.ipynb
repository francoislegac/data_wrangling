```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#tracer sinx
x = np.linspace(-2*np.pi, 2*np.pi, 100)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_title('f(x)=sin(x)');
```

```python
#if u want to add another line
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='sin')
ax.plot(x, np.cos(x), label='cos')
ax.legend(loc='upper center')
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.grid();
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='sin')
ax.plot(x, np.cos(x), label='cos')
ax.legend(loc='upper center')
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_ylim([0,1])
ax.set_yticks(np.linspace(0,1,3, endpoint=True))
ax.set_xlim([-2*np.pi, 0])

#x ticks
ax.set_xticks([-2*np.pi+k*np.pi for k in range(3)])
ax.set_xticklabels(['-2$\pi$', '$\pi/2$', '$0$'])
;
```

```python
#©hange the style of line
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), 'b', linewidth =2)
ax.plot(x, np.sin(x) + 1, 'r--', linewidth =2)
ax.plot(x, np.sin(x)-1, 'g-.', linewidth =2)
;
```

```python
#IMPORTANT 
#sharex=True, sharey=True
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(10, 5))  # 1 row, 2 columns

# First subplot
ax[0].plot(x, np.sin(x), linewidth=2, color='red')
ax[0].grid()
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].set_title("Sine curve")

# Second item
ax[1].plot(x, np.cos(x)*np.exp(-x/2), linewidth=2, color='green')
ax[1].grid()
ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
ax[1].set_title("Cos * exp curve");
```
